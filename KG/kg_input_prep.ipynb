```python
import pandas as pd
import re
from konlpy.tag import Kkma
```

```python
total = pd.read_csv('../kg_input.tsv', sep='\t', names=['re_tag', 'text', 'e1_tag', 'e2_tag'])
display(total.info())
display(total.head())
```

```python
e1_re = re.compile('<e1> .+ </e1>')
e2_re = re.compile('<e2> .+ </e2>')

e1_lst = []
e2_lst = []

for sent in total['text'].values:
    e1_lst.append(e1_re.findall(sent)[0].replace('<e1>', '').replace('</e1>', '').strip())
    e2_lst.append(e2_re.findall(sent)[0].replace('<e2>', '').replace('</e2>', '').strip())
    
total['e1'] = e1_lst
total['e2'] = e2_lst

total.head()
```

```python
total.to_csv('kg_input_required_info.tsv', sep='\t', encoding='utf-8', index=False)
```

```python
with open('stopwords.txt', encoding='utf-8') as f:
    stopwords = f.read()
    
stopwords = stopwords.split()
stopwords
```

```python
new_e1_lst = []
new_e2_lst = []

for e1 in total['e1'].values:
    e1 = e1.replace(",", "").replace("(", "").replace(")", "")
    
    for stopword in stopwords:
        if e1 not in ['모더나', '오한', '심한', '발열·두통·오한'] and e1.endswith(stopword):
            stop_len = len(stopword)
            e1 = e1[:-stop_len]
            break
    new_e1_lst.append(e1)
    
for e2 in total['e2'].values:
    e2 = e2.replace(",", "").replace("(", "").replace(")", "")
    
    for stopword in stopwords:
        if e2 not in ['모더나', '오한', '심한', '발열·두통·오한'] and e2.endswith(stopword):
            stop_len = len(stopword)
            e2 = e2[:-stop_len]
            break
    new_e2_lst.append(e2)
    
total['new_e1'] = new_e1_lst
total['new_e2'] = new_e2_lst

total.head()
```

```python
total.to_csv('kg_input_required_info_filtered.tsv', sep='\t', encoding='utf-8', index=False)
```
