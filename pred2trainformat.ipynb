## pred_out.txt 를 train.tsv 의 형식으로 바꾸기

```python
pred_sent_list = []
pred_tag_list = []

pred_list = [line.rstrip() for line in open('KoBERT-NER/sample_pred_out.txt', encoding = 'utf8')]
for sample in pred_list:
    sample_split = [word.replace('[', '').replace(']','').replace(':','') for word in sample.split(' ')]
    labels = ['UNK','VACC-B','VACC-I','SYMP-B','SYMP-I','DIS-B','DIS-I']
    word_list = []
    label_list = []
    for word in sample_split :
        if word.endswith(tuple(labels)):
            for label in labels:
                if word.endswith(label):
                    word_list.append(word.replace(label,''))
                    label_list.append(label) 
        else:
            label_list.append('O')
            word_list.append(word)

    pred_sent_list.append([' '.join(word_list)])
    pred_tag_list.append([' '.join(label_list)])
```

# merge tags

```python
from itertools import combinations

def merge_tags(tags):
        prior = False
        prior_tag = False
        chunk_dict, chunk_id = dict(), 0
        for i, tag in enumerate(tags):
            bio = tag[-1]
            if bio == "B":
                if prior:
                    chunk_dict[chunk_id]['end'] = i - 1
                chunk_id += 1
                chunk_dict[chunk_id] = {"entity_type": tag[:-2], "begin": i, "end": i}
                prior = True
                prior_tag = chunk_dict[chunk_id]['entity_type']  # 'SYMP'
                #print(prior_tag)
            elif bio == "I":
                if (prior_tag == False) or (tag[:-2] != prior_tag) : # 앞전의 태그가 O였거나 다른 태그였던 경우
                    #print(prior_tag, bio)
                    chunk_id +=1
                    chunk_dict[chunk_id] = {"entity_type": tag[:-2], "begin": i, "end": i}
                    prior = True
                    prior_tag =chunk_dict[chunk_id]['entity_type'] 
                else: # 앞전의 태그와 동일한 태그가 이어서 나온 경우
                    chunk_dict[chunk_id]['end'] = i 
            elif bio == "O":
                if prior:
                    chunk_dict[chunk_id]['end'] = i - 1
                prior = False
                prior_tag = False
        return chunk_dict
```

# to relation foramt

```python
def to_relation_format(sent_list, tag_list):
        result = []
        for sent, tags in zip(sent_list, tag_list):
            
            #print(sent,tags)
            chunk_dict = merge_tags(tags)
            #print(chunk_dict)
            if len(chunk_dict) < 2:  # 엔티티 갯수가 2개 이상일 때만 포함
                continue
            for subj, obj in combinations(chunk_dict, 2):
                #print(chunk_dict[subj], chunk_dict[obj])
                re_sent = sent.copy()
                subj, obj = chunk_dict[subj], chunk_dict[obj]
                subj_begin, subj_end, obj_begin, obj_end = subj['begin'], subj['end'], obj['begin'], obj['end']
                re_sent[subj_begin] = "<e1>" + re_sent[subj_begin]
                re_sent[subj_end] = re_sent[subj_end] + "</e1>"
                re_sent[obj_begin] = "<e2>" + re_sent[obj_begin]
                re_sent[obj_end] = re_sent[obj_end] + "</e2>"
                result.append((" ".join(re_sent), subj["entity_type"], obj["entity_type"]))
        return result
```
